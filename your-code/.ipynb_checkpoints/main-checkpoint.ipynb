````python
# your code here
Developers = []

#for full names 
result  = soup.find_all('h1', attrs = {'class' :'h3 lh-condensed'})
result  

for name in result:
    developer_name = name.get_text(separator=' ').strip()
    Developers.append(developer_name)

# Print the list of developers
print(Developers)
````

````python
#for nicknames
nicknames = []
for nick_name in soup.find_all('p', class_='f4 text-normal mb-1'):
    nickname = nick_name.find('a').text.strip()
    nicknames.append(nickname)

output = [f'{name} (@{nickname})' for name, nickname in zip(Developers, nicknames)]

for item in output:
    print(item)
````

#### 2. Display all the image links from Walt Disney wikipedia page.
Hint: use `.get()` to access information inside tags. Check out the documentation.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# your code here
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

img_links = soup.find_all('a', class_='mw-file-description')

image_urls = []

# Iterate over the class elements and get the 'href' attribute
for img_link in img_links:
    image_url = img_link.get('href')
    
    image_urls.append(image_url)

display(image_urls)
````

#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

language_elements = soup.find_all('a', id=lambda x: x and 'js-link-box' in x)

for lang in language_elements:
    language_name = lang.find('strong').get_text(strip=True)
    article_count = lang.find('bdi').get_text(strip=True)
    print(f'{language_name}: {article_count} articles')
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
response=requests.get(url)
response
````

````python
# your code here
#headers allow you to "pretend" to be a web browser
headers= {'Accept-Encoding':'gzip, deflate',
          'Accept-Language':'en-US,en;q=0.9',
          'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'}
````

````python
response = requests.get(url, headers = headers)
response.status_code
````

````python
tree = html.fromstring(response.content)
````

````python
#This will create a list of movies:
movies = tree.xpath('//h3[@class="ipc-title__text"]/text()')
print(movies)
````

````python
soup = BeautifulSoup(response.text, 'html.parser')

movies = soup.select('.ipc-metadata-list-summary-item')

Movie_data = []


for index in range(0, len(movies)):

    movie_title = movies[index].select_one('.ipc-title__text').text

    initial_release = movies[index].select_one('.sc-479faa3c-7.jXgjdT.cli-title-metadata span').text

    director_name = ""

    rating = float(movies[index].select_one('.ipc-rating-star').get_text().split()[0])

    Movie_data.append({"movie_title": movie_title, "initial_release": initial_release, "director": director_name, "rating": rating})

Movie_df = pd.DataFrame(Movie_data)
#to remove leading characters '1', '2', '3', '4', '5', '6', '7', '8', '9', '0', and '.' from the beginning of each string in the 'movie_title' column.
Movie_df['movie_title'] = Movie_df['movie_title'].str.lstrip('1234567890. ')

display(Movie_df)
#The element for the Director's has been disabled
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
response=requests.get(url)
response
````

````python
# your code here
Soup=BeautifulSoup(response.content,'html.parser')
Soup
````

````python
#Finds all HTML div elements with the class 'lister-item-content'
movies_ = Soup.find_all('div', class_='lister-item-content')

Top_movies = []

#Iterates through these div elements and extracts the movie title and summary
for movie in movies_:
    title_element = movie.find('h3', class_='lister-item-header').a
    movie_title = title_element.text.strip() if title_element else 'No title'

    summary_element = movie.find('p', class_='')
    summary = summary_element.text.strip() if summary_element else 'No summary'

    Top_movies.append({'movie name': movie_title, 'summary': summary})

Top_Movies = pd.DataFrame(Top_movies)

display(Top_Movies)


````
